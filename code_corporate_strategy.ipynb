```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define years and base revenue (in billion euros)
years = np.arange(2024, 2030)
base_revenue = 50  # Example starting revenue in 2024

# Growth rates for different scenarios
growth_rates = {
    "Pessimistic": 0.02,  # 2% annual growth
    "Neutral": 0.05,      # 5% annual growth
    "Optimistic": 0.08    # 8% annual growth
}

# Generate projections
projections = {scenario: [base_revenue * (1 + rate) ** (year - 2024)
                          for year in years]
               for scenario, rate in growth_rates.items()}

# Convert to DataFrame
df = pd.DataFrame(projections, index=years)

# Plot with colors
sns.set_style("whitegrid")
plt.figure(figsize=(10, 5))
colors = {"Pessimistic": "red", "Neutral": "blue", "Optimistic": "green"}

for scenario in df.columns:
    plt.plot(df.index, df[scenario], marker='o', linestyle='-',
             color=colors[scenario], label=scenario)

# Labels and title
plt.xlabel("Year", fontsize=12)
plt.ylabel("Revenue (Billion )", fontsize=12)
plt.title("IKEA Revenue Forecast (2024-2029)", fontsize=14, fontweight="bold")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Define historical revenue and profit margin data (example values in billion EUR)
data = {
    'Year': np.arange(2015, 2024),
    'Revenue': [32.7, 35.1, 36.3, 38.8, 41.3, 42.0, 41.9, 44.6, 47.6],  # Example IKEA revenue data
    'Profit Margin': [0.08, 0.09, 0.10, 0.095, 0.11, 0.10, 0.105, 0.11, 0.115]  # Example profit margins
}

# Create DataFrame
df = pd.DataFrame(data)

# Define projection years
future_years = np.arange(2024, 2030)

# Growth rate assumptions
neutral_growth = 0.04  # 4% annual growth
pessimistic_growth = 0.02  # 2% annual growth
optimistic_growth = 0.06  # 6% annual growth

# Profit margin assumptions
neutral_margin = 0.11  # 11% profit margin
pessimistic_margin = 0.09  # 9% profit margin
optimistic_margin = 0.12  # 12% profit margin

# Get last known revenue and profit margin
last_revenue = df['Revenue'].iloc[-1]
last_profit_margin = df['Profit Margin'].iloc[-1]

# Generate forecasts
neutral_forecast = [last_revenue * (1 + neutral_growth) ** (i + 1) for i in range(len(future_years))]
pessimistic_forecast = [last_revenue * (1 + pessimistic_growth) ** (i + 1) for i in range(len(future_years))]
optimistic_forecast = [last_revenue * (1 + optimistic_growth) ** (i + 1) for i in range(len(future_years))]

# Create profit forecasts
neutral_profit = [rev * neutral_margin for rev in neutral_forecast]
pessimistic_profit = [rev * pessimistic_margin for rev in pessimistic_forecast]
optimistic_profit = [rev * optimistic_margin for rev in optimistic_forecast]

# Create forecast DataFrame
forecast_df = pd.DataFrame({
    'Year': future_years,
    'Pessimistic Revenue': pessimistic_forecast,
    'Neutral Revenue': neutral_forecast,
    'Optimistic Revenue': optimistic_forecast,
    'Pessimistic Profit': pessimistic_profit,
    'Neutral Profit': neutral_profit,
    'Optimistic Profit': optimistic_profit
})

# Define revenue sources (example proportions)
sources = ['Furniture', 'Kitchen', 'Decoration', 'Services']
proportions = [0.5, 0.2, 0.2, 0.1]  # 50%, 20%, 20%, 10%

# Create detailed table for 2024 and 2029
forecast_2024 = {'Year': 2024, 'Total Revenue': last_revenue, 'Total Profit': last_revenue * last_profit_margin}
forecast_2029 = {
    'Year': 2029,
    'Pessimistic Revenue': pessimistic_forecast[-1],
    'Neutral Revenue': neutral_forecast[-1],
    'Optimistic Revenue': optimistic_forecast[-1],
    'Pessimistic Profit': pessimistic_profit[-1],
    'Neutral Profit': neutral_profit[-1],
    'Optimistic Profit': optimistic_profit[-1]
}

# Create detailed breakdown
breakdown = []
for source, proportion in zip(sources, proportions):
    breakdown.append({
        'Source': source,
        '2024 Revenue': last_revenue * proportion,
        '2024 Profit': (last_revenue * proportion) * last_profit_margin,
        '2029 Pessimistic Revenue': forecast_2029['Pessimistic Revenue'] * proportion,
        '2029 Pessimistic Profit': forecast_2029['Pessimistic Profit'] * proportion,
        '2029 Neutral Revenue': forecast_2029['Neutral Revenue'] * proportion,
        '2029 Neutral Profit': forecast_2029['Neutral Profit'] * proportion,
        '2029 Optimistic Revenue': forecast_2029['Optimistic Revenue'] * proportion,
        '2029 Optimistic Profit': forecast_2029['Optimistic Profit'] * proportion
    })

breakdown_df = pd.DataFrame(breakdown)

# Calculate total gains
total_gains = {
    'Source': 'Total Gains',
    '2024 Revenue': last_revenue,
    '2024 Profit': last_revenue * last_profit_margin,
    '2029 Pessimistic Revenue': forecast_2029['Pessimistic Revenue'],
    '2029 Pessimistic Profit': forecast_2029['Pessimistic Profit'],
    '2029 Neutral Revenue': forecast_2029['Neutral Revenue'],
    '2029 Neutral Profit': forecast_2029['Neutral Profit'],
    '2029 Optimistic Revenue': forecast_2029['Optimistic Revenue'],
    '2029 Optimistic Profit': forecast_2029['Optimistic Profit']
}
breakdown_df = pd.concat([breakdown_df, pd.DataFrame([total_gains])], ignore_index=True)

# Print table
print(breakdown_df.to_string(index=False))

# Plot the forecasts
plt.figure(figsize=(10, 5))
plt.plot(df['Year'], df['Revenue'], marker='o', label='Historical Revenue', color='black')
plt.plot(forecast_df['Year'], forecast_df['Pessimistic Revenue'], marker='o', linestyle='dashed', label='Pessimistic Revenue (2%)', color='red')
plt.plot(forecast_df['Year'], forecast_df['Neutral Revenue'], marker='o', linestyle='dashed', label='Neutral Revenue (4%)', color='blue')
plt.plot(forecast_df['Year'], forecast_df['Optimistic Revenue'], marker='o', linestyle='dashed', label='Optimistic Revenue (6%)', color='green')

plt.xlabel('Year')
plt.ylabel('Revenue (Billion EUR)')
plt.title('IKEA Revenue Forecast (2024-2029)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Base year data (2024 actual figures)
base_revenue = 34313  # in million euros
base_gross_profit = 5490  # in million euros
base_service_share = 0.03  # 3% of revenue
base_service_margin = 0.20  # 20% gross margin for services
base_product_margin = 0.16  # 16% gross margin for products

# Growth assumptions
growth_rates = {
    "Pessimistic": 0.02,
    "Neutral": 0.04,
    "Optimistic": 0.06
}

# Service sector expansion assumptions
service_growth = {
    "Pessimistic": (0.03, 0.06, 0.25),  # (start share, end share, final margin)
    "Neutral": (0.03, 0.09, 0.30),
    "Optimistic": (0.03, 0.12, 0.35)
}

# Data storage
data = []

# Forecast for 2025-2029
for year in range(2025, 2030):
    for scenario in ["Pessimistic", "Neutral", "Optimistic"]:
        previous_revenue = base_revenue if year == 2025 else data[-3]["Total Revenue"]
        growth_rate = growth_rates[scenario]
        new_revenue = previous_revenue * (1 + growth_rate)

        # Service revenue projections
        start_share, end_share, final_margin = service_growth[scenario]
        service_share = start_share + (end_share - start_share) * (year - 2024) / 5  # Linear increase
        service_revenue = new_revenue * service_share
        product_revenue = new_revenue - service_revenue

        # Profit calculation
        service_margin = base_service_margin + (final_margin - base_service_margin) * (year - 2024) / 5  # Linear increase
        service_profit = service_revenue * service_margin
        product_profit = product_revenue * base_product_margin
        total_profit = service_profit + product_profit

        data.append({
            "Year": year,
            "Scenario": scenario,
            "Total Revenue": round(new_revenue, 2),
            "Product Revenue": round(product_revenue, 2),
            "Service Revenue": round(service_revenue, 2),
            "Product Profit": round(product_profit, 2),
            "Service Profit": round(service_profit, 2),
            "Total Profit": round(total_profit, 2)
        })

# Convert to DataFrame
df = pd.DataFrame(data)
print(df)
```
